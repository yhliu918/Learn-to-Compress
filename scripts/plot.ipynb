```python
inverse=np.loadtxt('../build/nohup_inverse_2000000.txt')
piecewise=[]
For = []
inverse_score =[]
for i in range(len(inverse)):
    if i %3 ==0:
        inverse_score.append(inverse[i])
    elif i%3 ==1:
        piecewise.append(inverse[i])
    elif i%3 ==2:
        For.append(inverse[i])
```

```python
dist = []
for  rate1, rate2 in zip( piecewise, For):
    dist.append(abs(rate1-rate2))
point = inverse_score[dist.index(min(dist))]
cr = For[dist.index(min(dist))]
plt.figure(figsize=(15,10))

plt.plot(inverse_score,piecewise, color='teal',label = 'Linear fix',linewidth=4)
plt.plot(inverse_score,For,color = 'yellowgreen', label = 'FOR',linewidth=4)
plt.scatter(point,cr,c='slategrey',s=100)
plt.text(point+0.20, cr+1, "("+str(round(point,2))+', '+str(round(cr,2))+")", fontsize=25, color = "slategrey", style = "italic", weight = "light", verticalalignment='center', horizontalalignment='right', rotation=0)
plt.ylabel('Compression ratio', fontdict={ 'size'   : 30})
plt.xlabel('$\mathcal{P}_{inv}$', fontdict={ 'size'   : 30})
plt.legend(prop={ 'size'   : 30})
plt.yticks(size = 22)
plt.xticks(size = 22)
plt.show()
```
