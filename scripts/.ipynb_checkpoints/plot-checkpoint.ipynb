```python
dd=np.loadtxt('../data/books_200M_uint32.txt')
```

```python
delta =[]
for i in range(len(dd)-1):
    delta.append(dd[i+1]-dd[i])
```

```python
ff='books_fanout.out'
seg =[]
file = open(ff)
while 1:
    line = file.readline()
    line =line.split(' ')
    if len(line)<3:
        break
        
    seg.append(int(line[2]))
```
